Ahora abriremos un editor de texto `vim 123` o `nano 123`


ya en el editor escribir, las lineas correspondientes al disco f se comentan porque se realizará otro procedimiento alternativo


***
```
#!/bin/bash
#Estos comandos Establecen que van a ser discos msdos master boot record 
parted /dev/sdb mklabel msdos
parted /dev/sdc mklabel msdos
parted /dev/sdd mklabel msdos
parted /dev/sde mklabel msdos
#parted /dev/sdf mklabel msdos

#Estos comandos crean una particion primaria y le hacen un set a lvm
parted -s /dev/sdb unit mib mkpart primary 1 100% set 1 lvm on
parted -s /dev/sdc unit mib mkpart primary 1 100% set 1 lvm on
parted -s /dev/sdd unit mib mkpart primary 1 100% set 1 lvm on
parted -s /dev/sde unit mib mkpart primary 1 100% set 1 lvm on
#parted -s /dev/sdf unit mib mkpart primary 1 100% set 1 lvm on

# Solo si es RHEL8
#Este comando le dice al kernel que reconozca esto nuevo
#udevadm settle

# De lo contrario ejecute
partprobe -s
```
***

guardamos los cambios y salimos

Lo anterior se puede hacer el con el comando **fdisk** mediante lo siguiente sobre el disco sdf

`fdisk /dev/sdf`

ya dentro de lo que pregunta **fdisk** hay que decirle los siguientes comandos

`n`	para crear una nueva particion
`p`	porque la partición será primaria
`1`	se elige el primer sector por defecto que es el 2048
    y presionamos enter para aceptar el último sector

luego le damos los siguientes comandos
`t`	para cambiar el tipo de la particion a lvm
`L`	para mostrar la lista de códigos que admite
`8e`escribimos 8e que es linux lvm
`p`	Una vez hecho esto hacemos el print
`w` escribimos w para guardar los cambios

Una vez que los cambios se escriben nos regresa al bash, en donde ejecutamos 
`partprobe -s`

**Nota:** El siguiente comando se usa únicamente si trabajamos con redhat `udevadm settle`

***

Practicamente siempre vamos a ejecutar el script para preparar los discos en lvm, en muchas ocasiones se trabaja con el disco completo, no se trabaja con una partición del disco,  pero vamos a seguir las buenas prácticas y vamos a tomar el disco creando una particion solamente y de allí entonces vamos a seguir.

**En cada practica habra que ejecutar este script, dependiendo del número de discos a utilizar. Ya que en los demás laboratorios, vamos a regresar al estado anterior de la máquina virtual (5 discos sin formato) por lo que hay que regresar a ese estado anterior y despues ejecutar el script.**

Para regresar al estado anterior hay que apagar la máquina virtual `poweroff` y con la máquina apagada, regresar al estado anterior habiendo seleccionado el estado **5 discos sin formato**

![02](04_img/02.png)

Quitar la marca de la casilla **Crear instantánea del estado actual de la máquina** (Create a snapshot of the current machine state) ya que no nos interesa conservar el estado actual en que se encuentre 

![03](04_img/03.png)

Una vez restaurado el estado, iniciar la máquina en el estado actual (el estado actual ahora es el estado en el que se realizó la instantánea con los 5 discos sin formatear)

![04](04_img/04.png)

Una vez iniciada la máquina virtual, conectarse a nuevamente `ssh -p 2222 alumno@127.0.0.1` y ejecutar el comando `fdisk -l` y comprobar que ahora los discos se encuentran sin formatear.
