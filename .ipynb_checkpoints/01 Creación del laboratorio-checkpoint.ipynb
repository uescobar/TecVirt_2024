# Creación del Laboratorio


## Software

Requerimientos de software

* [VirtualBox] (https://www.virtualbox.org/)
* [Debian] (https://cdimage.debian.org/debian-cd/current/amd64/iso-cd/debian-12.4.0-amd64-netinst.iso)




## Servidor
### Creacion de maquina virtual

* Debian 12
* 1 GB de RAM (recomendable 2GB)
* 1 CPU (recomendable 2GB)
* 1 Disco primario de 10 GB
* 5 Discos secundarios de 1 GB (recomendable 5 GB)


Abrir VirtualBox

![01](01_img/01.png)


En virtualbox Crear una Máquina Virtual de nombre **lvm1** (marcar la casilla Skip Unattended Instalation)

![02](01_img/02.png)


Agregar 2 GB de memoria RAM y 2 procesadores

![03](01_img/03.png)


Seleccionar la Capacidad de Disco Duro de 10 GB

![04](01_img/04.png)


Revisar y en su caso aceptar el resumen de la creación de la máquina virtual (aceptar al dar clic en el botón **finish**)

![05](01_img/05.png)

Posteriormente, encender la maquina virtual **lvm1** con un **inicio normal** botón **inicio o start** (con la flecha verde)

![06](01_img/06.png)


## Instalación de Software

Una vez inicada la máquina virtual vamos a ejecutar los siguientes pasos


Seleccionar instalación gráfica (Graphical Install)

![07](01_img/07.png)

Seleccionar el idioma español

![08](01_img/08.png)

Seleccionar México

![09](01_img/09.png)

Seleccionar Latinoamericano

![10](01_img/10.png)

Comienza el primer proceso de instalación y hay que esperar un poco

![11](01_img/11.png)

Establecer el nombre de la máquina **debian12lvm1**

![12](01_img/12.png)

Establecer el dominio de la máquina **labdebian.com**

![13](01_img/13.png)

Establecer la contraseña del usuario **root** de la máquina, la contraseña es **uiop00**

![14](01_img/14.png)

Establecer el nombre completo del usuario y escribir **alumno**

![15](01_img/15.png)

Establecer el nombre de usuario y escribir **alumno**

![16](01_img/16.png)

Establecer la contraseña del usuario **alumno** y escribir **uiop00**

![17](01_img/17.png)

Se ejecuta un segundo proceso de instalación

![18](01_img/18.png)

Seleccionamos hora central

![19](01_img/19.png)

Se inicia otro proceso de instalación

![20](01_img/20.png)

Seleccionamos la opción **Utilizar todo el disco**

![21](01_img/21.png)

Seleccionar el único disco que aparece allí

![22](01_img/22.png)

Seleccionar **todos los ficheros en una partición (recomendada para novatos)**

![23](01_img/23.png)

Se inicia otro proceso y muestra la siguiente pantalla, verificar que esté seleccionado la opción **Finalizar particionado y escribir los cambios en el disco**

![24](01_img/24.png)

Marcar **Si** a la pregunta de **¿Desea escribir los cambios en los discos**

![25](01_img/25.png)

Se inicia un proceso de instalación y esperamos unos minutos a que se realize este proceso

![26](01_img/26.png)

A la pregunta de **¿Desea analizar métodos de instalación adicionales** debemos de elegir la opción **No**

![27](01_img/27.png)

Al elegir **Configurar el gestor de paquetes** elegir **Estados Unidos**

![28](01_img/28.png)

Elegir la opción **deb.debian.org**

![29](01_img/29.png)

Dejar en blanco la opción de proxy

![30](01_img/30.png)

Se realiza el proceso de configuración de apt

![31](01_img/31.png)

A la pregunta de  **¿Desea participar en la encuesta sobre el uso de paquetes?** debemos responder que no

![32](01_img/32.png)

Se realiza otro proceso de seleccion de paquetes

![33](01_img/33.png)

Se seleccionan las opciones **SSH server** y **Utilidades estánadar del sistema**

![34](01_img/34.png)

Se instala el software seleccionado
