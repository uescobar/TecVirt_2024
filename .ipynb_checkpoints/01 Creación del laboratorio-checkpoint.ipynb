## Configuración de la máquina virtual

loguearse como el usuario root

![41](01_img/41.png)


A continuación se hará la descripción con puros comandos, sobre la configuración,

### Configuración el usuario alumno

Lo primero es instalar sudo mediante el comando `apt install sudo`

Hacer que el usuario **alumno** pueda tomar el rol de super usuario agregandolo al grupo de super usuarios mediante el comando `usermod -aG sudo alumno` lo anterior es debido a que el usuario root tiene bloqueado el acceso mediante ssh, y requeriremos acceder como el usuario alumno para ejecutar comandos que modifiquen el sistema (de forma opcional se puede permitir el acceso mediante ssh con el usuario root deshabilitando la restricción en el archivo **/etc/ssh/sshd_config**).

### Configuración de la red 

Ejecutar el comando `ip a` para corroborar la dirección ip, con lo cual se muestra en nuestro caso **10.0.2.15/24** en el dispositivo **enp0s3**

Ejecutar el comando `ip route` para mostrar información sobre la puerta de enlace definida por defecto, que en nuestro caso, es la dirección **10.0.2.2**

Ahora debemos poner la direccion fija, ya que esta dirección del servidor DNS la obtuvo por DHCP la debemos cambiar, lo hacemos con el comando `cat /etc/resolv.conf`, que en nuetro caso se muestra como **nameserver 192.168.2.207** y debe cambiarlo por **nameserver 8.8.8.8** 

La configutación anterior muestra que **no podremos acceder desde la red local del anfitrion al host**, por lo que debemos modificar la red de la maquina virtual, ya que sino no podremos acceder a ella, lo cual lo podemos hacer de varias formas, una de ellas es sacándola de la red privada virtual, la segunda es haciendo mediante el reenvío de puertos. Realizemoslo mediante el reenvío de puertos

En la máquina virtual, como usuario root, ejecutar el comando `cd /etc/network/` para ir al directorio en donde se encuentra el archivo de configuración.

Ejecutar el siguiente comando para crear una copia del archivo antes de modificarlo `cp interfaces interfaces.old`

Ejecutar el comando `nano interfaces` para editar el archivo de configuración y dejar el contenido de la siguiente manera.

**intarfaces**

***
```                              
# This file describes the network interfaces available on your system
# and how to activate them. For more information, see interfaces(5).

source /etc/network/interfaces.d/*

# The loopback network interface
auto lo
iface lo inet loopback

# The primary network interface
allow-hotplug enp0s3
iface enp0s3 inet static
        address 10.0.2.15
        netmask 255.255.255.0
        gateway 10.0.2.2


```
***

reiniciar la máquina virtual con el comando `reboot`

Una vez iniciada la máquina virtual, verificar que tenga la dirección ip asignada con el comando `ip a` y con el comando `ping 8.8.8.8 -c 5` debe verificar que tenga acceso a internet.

Una vez realizado lo anterior, en el panel de virtualbox, en la configuración de la máquina virtual **lvm1** en el apartado de red (network), expandir la sección **Avanzado** y dar clic en el Botón **Reenvio de puertos** (Port Forwading)

![42](01_img/42.png)

Y Asignar una nueva regla como se muestra a continuación

![43](01_img/43.png)

| Nombre   | Protocol| Host IP  | Host Port | Guest IP  | Guest Port |
| -------- | ------- | -------- | -------   | --------  | -------    |
| Rule1    | TCP     | 127.0.0.1|    2222   | 10.0.2.15 |     22     |


Ahora desde la terminal (**en Windows desde la terminal de gitBash**) ejecutar el comando

`ssh -p 2222 alumno@127.0.0.1`, solicitando confirmar la identidad del servidor y deberá escribir **yes**, posteriormente, la contraseña del usuario **alumno** y si todo esta correcto se mostrará el resultado parecido a 

![44](01_img/44.png)


Con lo anterior, la configuración básica de la máquina virtual, queda concluida y solo resta apagarla para cuando se use la siguiente ocasión.

Apagar la máquina virtual mediante el comando `sudo poweroff`, proporcionar la clave del usuario **alumno** y la maquina virtual debe apagarse.
