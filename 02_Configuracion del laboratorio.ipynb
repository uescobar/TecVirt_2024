Ahora tenemos que agragar mas discos duros, pero primero debemos apagar la máquina virtual `poweroff`


Ahora desde la interfaz de virtualbox en el apartado de **Configuración** (Settings) nos dirigimos a la sección de **almacenamiento** (Storage)

![01](02_img/01.png)

Y cambiamos el valor de el atributo **Port count** a **6**

![02](02_img/02.png)

Y procedemos a agregar discos duros dando clic en el boton que tiene le ícono de simbolo mas sobre un disco duro, el último botón a la derecha de donde dice **Controller Sata** y se abre el cuadro de dialogo que se muestra a continuación

![03](02_img/03.png)

Procedemos a dar clic en el boton **Create**, debemos seleccionar **VHD**

![04](02_img/04.png)

Seleccionamos **Fixed Size**

![05](02_img/05.png)

Ahora seleccionamos el tamaño de 5 GB y verificamos que el nombre del archivo lleve el sufijo **_raid1** y dar clic en el botón crear.

![06](02_img/06.png)

Repetir el procedimiento hasta tener 5 discos de 5 GB cada uno.

![07](02_img/07.png)

Ahora comenzamos seleccionando uno a uno cada disco duro, en orden ascendente comenzando con el disco raid1, y lo vamos agregando a la máquina virtual, hata tener el siguiente resultado

![08](02_img/08.png)
